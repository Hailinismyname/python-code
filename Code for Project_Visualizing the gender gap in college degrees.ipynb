```python
import matplotlib.pyplot as plt
import pandas as pd
# reading file, setting up three categories that need to generate line charts using for loop,
# setting the line color
women_grads=pd.read_csv('percent-bachelors-degrees-women-usa.csv') 
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
women_grads_line=(0/255, 107/255, 164/255)
men_grads_line=(255/255, 128/288, 14/255)

#create subplots, using for loop create line charts with x-axis 'year' and y-axis stem_cats,
#set color as women_grads_line, add label and line width, set x and y range 
#remove the spine of plots, disable the tick
#locate legend at certain plot
#hiding x-axis for most of plots, only enable x-axis for the bottommost plot
#make only 0, 50 100 points visible for y-axis
#adding a horizontal line with RGB setup, transparency 0.5
fig=plt.figure(figsize=(14,26))
for each in range(0,6):
    ax=fig.add_subplot(6,3,3*each+1)
    ax.plot(women_grads['Year'], women_grads[stem_cats[each]], c=women_grads_line, linewidth=3)
    ax.plot(women_grads['Year'], 100-women_grads[stem_cats[each]], c=men_grads_line, linewidth=3)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    for key,value in ax.spines.items():
        value.set_visible(False)
   
    if each == 0:
        ax.text(2005,80,'Men')
        ax.text(2002,15,'Women')
    elif each == 5:
        ax.text(2005,62, 'Men')
        ax.text(2001,32, 'Women')
    ax.set_title(stem_cats[each])
    ax.tick_params(top='off', left='off', right='off', bottom='off', labelbottom='off')
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.5)
    ax.set_yticks([0,50,100])
ax.tick_params(labelbottom='on')

for each in range(0,5):
    ax=fig.add_subplot(6,3,3*each+2)
    ax.plot(women_grads['Year'], women_grads[lib_arts_cats[each]], c=women_grads_line, linewidth=3)
    ax.plot(women_grads['Year'], 100-women_grads[lib_arts_cats[each]], c=men_grads_line, linewidth=3)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    for key,value in ax.spines.items():
        value.set_visible(False)
    ax.tick_params(top='off', left='off', right='off', bottom='off', labelbottom='off')
    if each == 0:
        ax.text(2005,75,'Men')
        ax.text(2002,20,'Women')
    ax.set_title(lib_arts_cats[each])
    ax.set_yticks([0,50,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.5)
ax.tick_params(labelbottom='on')

for each in range(0,6):
    ax=fig.add_subplot(6,3,3*each+3)
    ax.plot(women_grads['Year'], women_grads[other_cats[each]], c=women_grads_line, linewidth=3)
    ax.plot(women_grads['Year'], 100-women_grads[other_cats[each]], c=men_grads_line, linewidth=3)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    for key,value in ax.spines.items():
        value.set_visible(False)
    ax.tick_params(top='off', left='off', right='off', bottom='off', labelbottom='off')
    if each == 0:
        ax.text(2005,87,'Men')
        ax.text(2002,8,'Women')
    elif each == 5:
        ax.text(2005,60,'Men')
        ax.text(2002,35,'Women')
    ax.set_title(other_cats[each]) 
    ax.set_yticks([0,50,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.5)

ax.tick_params(labelbottom='on')
plt.savefig('biology_degrees.png')
plt.show()
```
