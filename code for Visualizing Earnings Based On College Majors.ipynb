```python
import matplotlib.pyplot as plt
import pandas as pd
# part time, full time 和 low_wage_jobs的关系
# bar plot

# 那一个major category 接到工作的毕业生最多(employed)
# 用employed排序

# men, women各占employed的比例，在高薪专业的比例
# 饼图

# 那些高薪是男人专有，哪些是女人专有
# histograms

# 热门专业的人数比较。
# bar plot


# read the file and check the header, structure, summary is a table 
# introduces the basic numeric info
%matplotlib inline
recent_grade=pd.read_csv('recent-grads.csv')
firsttwo_line=recent_grade.head(2)
tail_row=recent_grade.tail(0)
summary = recent_grade.describe()

# remove the missing item and assign to raw_count 
raw_data_count = recent_grade.index
recent_grade = recent_grade.dropna()
raw_count = recent_grade.index

```

```python
# using Pandas drows the plot
ax1 = recent_grade.plot(x='Sample_size', y='Median', kind='scatter')
ax1.set_title('Median vs. Sample_size')

ax2 = recent_grade.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax2.set_title('Unemployment_rate vs. Sample_size')

ax3 = recent_grade.plot(x='Full_time', y='Median', kind='scatter')
ax3.set_title('Full_time vs. Median')

ax4 = recent_grade.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax4.set_title('ShareWomen vs. Unemployment_rate')

ax4 = recent_grade.plot(x='Men', y='Median', kind='scatter')
ax4.set_title('Men vs. Median')

ax5 = recent_grade.plot(x='Women', y='Median', kind='scatter')
ax5.set_title('Women vs. Median')

plt.show()
```

```python
# generating the histograms with pandas
lst=['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']
for i in lst:
    recent_grade[i].hist(bins=30)
    plt.title(i)
    plt.show()

```

```python
from pandas.tools.plotting import scatter_matrix 
scatter_matrix(recent_grade[['Rank', 'Full_time']], figsize=(10,10))
scatter_matrix(recent_grade[['Sample_size', 'Median','Unemployment_rate']], figsize=(10,10))

#第一幅图的第三个scatter 可以看出来，人最多的专业排收入排行在第大概75名左右，第二名就跌到了150名左右。
#大部分的专业能找到full time的都很少，不过还是要看总专业人数才能定论
scatter_matrix(recent_grade[['Total', 'Full_time']], figsize=(10,10))
#基本上呈正比例函数，人数越多的专业拿到fulltime人也就越多。很可能专业越热门
```

```python
col=['Women','Men']
recent_grade[col][:10].plot(kind='bar',x='Major', figsize=(12,16), title='Women vs. Men in first ten popular major',
                          table=False, grid=True, ylim=(0,250000) )
#recent_grade.plot.bar(x='Major', y=['Women','Men'],figsize=(12,16), title='Women vs. Men in first five popular major',
#                          table=False, grid=True, ylim=(0,250000))
#plt.show()
#recent_grade.plot.bar()更方便清晰一点，因为可以直接设置下x,y的label

#sorting the dataset to decending order
index=recent_grade.index
index_list=index.tolist()
sorted_index=sorted(index_list, reverse=True)
sorted_grade=recent_grade.reindex(index=sorted_index)

new_grade=recent_grade.sort_index(ascending=False)
sorted_grade.head(4)

new_grade[:10].plot.bar(x='Major', y=['Women','Men'],figsize=(12,16), title='Women vs. Men in first ten lowest pay',
                          table=False, grid=True)
new_grade[:10].plot.bar(x='Major', y='Unemployment_rate',figsize=(12,16), title='Unemployment rate in first ten lowest pay',
                          table=False, grid=True)
plt.show()
```
